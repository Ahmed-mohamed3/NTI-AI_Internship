```python
from tensorflow import keras
from keras.layers import Input, Dense, Dropout,Conv2D, MaxPooling2D, Flatten
from tensorflow.keras import  datasets
from matplotlib import pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = datasets.cifar10.load_data()
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#Normalize image pixel values.
x_train = x_train/255
x_test = x_test/255
```

```python
model=keras.Sequential() #Create a network sequence.

model.add(Input(shape=(32,32,3)))
model.add(Conv2D(50, (3, 3), activation='relu'))
#model.add(Conv2D(32, (3, 3), activation='relu',input_shape = (32,32,3)))
model.add(Conv2D(40, (3, 3), activation='linear', padding = 'same'))
model.add(Conv2D(30, (3, 3), activation='linear', padding = 'valid'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(28, (3, 3), activation='relu',padding = 'valid'))

model.add(Conv2D(43, (3, 3), activation='relu',padding = 'valid'))
model.add(Conv2D(30, (3, 3), activation='linear',padding = 'same'))


model.add(MaxPooling2D((2, 2)))

model.add(Flatten())
model.add(Dense(60,activation = 'relu'))
model.add(Dense(40,activation = 'relu'))
model.add(Dense(30,activation = 'relu'))
model.add(Dense(20,activation = 'linear'))
model.add(Dense(10,activation = 'softmax'))

model.summary()
```

```python
model.compile(optimizer='Adam',loss="sparse_categorical_crossentropy",metrics=['accuracy']) 
results= model.fit(x_train,y_train,epochs=30,batch_size=128,validation_data=(x_test, y_test))

```

```python
score = model.evaluate(x_test, y_test)
```

```python
y_pred=model.predict(x_test)
y_pred[0:3]
```

```python
import numpy as np
from numpy import argmax
y_pred1=argmax(y_pred,axis=1)
y_pred1[0:3]
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print(confusion_matrix(y_pred1,y_test))
print(accuracy_score(y_pred1,y_test))
```
