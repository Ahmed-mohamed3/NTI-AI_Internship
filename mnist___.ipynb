```python
import pandas as pd
import keras 
import numpy as np
from keras.models import *
from keras.layers import Dense
import matplotlib.pyplot as plt
from keras.datasets import mnist
import tensorflow.keras.layers as layers
```

```python
import tensorflow as tf
mnist = tf.keras.datasets.mnist
(x_train,y_train) , (x_test , y_test) = mnist.load_data()
print(x_train.shape)
print(y_train.shape)
```

```python
#Use the output of the previous layer as the input of the next layer.
x = tf.keras.Input(shape=(32,))
h1 = layers.Dense(32, activation='relu')(x)
h2 = layers.Dense(32, activation='relu')(h1)
y = layers.Dense(10, activation='softmax')(h2)
model_sample_2 = tf.keras.models.Model(x, y)
#Print model information.
model_sample_2.summary()
```

```python
np.unique(y_train)
```

```python
x_train_final = x_train.reshape((-1,784))
x_train_final.shape
```

```python
x_test_final = x_test.reshape((-1,784))
```

```python
# normlize the data
x_train_final = x_train_final/255
x_test_final  = x_test_final/255
```

```python
model = Sequential()
model.add(Dense(10,input_shape=(784,), activation="softmax"))
```

```python
model.compile(optimizer='sgd',loss= 'sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
#results = model.fit(x=x_train, y=y_train,shuffle=True,epochs=10)
results=model.fit(
  x=x_train_final,
  y=y_train,
  shuffle=True,
  epochs=30,
  validation_data=(x_test_final,y_test)
)
```

```python
import matplotlib.pyplot as plt
plt.plot(results.history['loss'])
plt.plot(results.history['val_loss'])
plt.legend(['Training','Vaildation'])
plt.title('Training and Validation Losses')
plt.xlabel('epoches')
plt.ylabel('losses')
```

```python
import matplotlib.pyplot as plt
plt.plot(results.history['accuracy'])
plt.plot(results.history['val_accuracy'])
plt.legend(['Training','Vaildation'])
plt.title('Training and Validation Accuracy')
plt.xlabel('epoches')
plt.ylabel('accuracy')
```
